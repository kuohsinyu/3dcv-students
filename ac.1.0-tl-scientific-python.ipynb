````python
import numpy as np
from skimage import io, color
from skimage.util import img_as_ubyte
import matplotlib.pyplot as plt

# TODO: 1. Load the image in ./data/pepo.jpg from your disk and store it in a variable named image
image = io.imread('./data/pepo.jpg')

# TODO: 2. Store and print the shape of the image and the value range (min and max values)
rgb_image_shape = image.shape
rgb_min = np.min(image)
rgb_max = np.max(image)

print(f'RGB image shape: {rgb_image_shape}')
print(f'RGB image min: {rgb_min}')
print(f'RGB image max: {rgb_max}')

# TODO: 3. Convert it to gray scale and save it in a variable named image_gray
image_gray = color.rgb2gray(image)

# TODO: 4. Store and print the shape of the grayscale image and the value range (min and max values)
grayscale_image_shape = image_gray.shape
grayscale_min = np.min(image_gray)
grayscale_max = np.max(image_gray)

print(f'Grayscale image shape: {grayscale_image_shape}')
print(f'Grayscale image min: {grayscale_min}')
print(f'Grayscale image max: {grayscale_max}')

# TODO: 5. Save the grayscale image back to disk in ./data/pepo_gray.jpg
io.imsave('./data/pepo_gray.jpg', img_as_ubyte(image_gray))
````

## Task 3

Plot the color histogram of the rgb image.

Reference: https://matplotlib.org/users/index.html

````python
import matplotlib.pyplot as plt
from skimage import exposure

def plot_color_histogram(image):
    '''
    Plot the histogram of each color channel into a single plot.
    Take a look at the scikit-image submodule skimage.exposure

    image -- RGB image
    '''
    # 確保影像是浮點或整數皆可
    colors = ('red', 'green', 'blue')
    channel_labels = ('R', 'G', 'B')

    plt.figure(figsize=(8, 5))

    for i, color_name in enumerate(colors):
        # 使用 skimage.exposure.histogram 計算每個通道的直方圖
        hist, bins = exposure.histogram(image[..., i])
        plt.plot(bins, hist, color=color_name, label=channel_labels[i])

    plt.title('Color Histogram (RGB Channels)')
    plt.xlabel('Pixel intensity')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# 呼叫函式繪圖
plot_color_histogram(image)
````

## Task 4

Convert the RGB image into the HSV color space. Rotate the colors in the HSV color space, convert it back to the RGB color space and plot it together with its histogram. For this task, you are only allowed to use NumPy.

More information about the HSV color space: https://en.wikipedia.org/wiki/HSL_and_HSV

````python
import numpy as np
import matplotlib.pyplot as plt

# ---------- RGB -> HSV（純 NumPy） ----------
def rgb2hsv(image):
    '''
    Convert the rgb image into hsv color space and return it.
    image -- RGB image, shape (H, W, 3), dtype uint8 or float.
    Returns float HSV in [0,1], shape (H, W, 3).
    '''
    img = image.astype(np.float64)
    if img.max() > 1.0:  # 常見的 uint8 圖片
        img = img / 255.0

    r, g, b = img[..., 0], img[..., 1], img[..., 2]
    maxc = np.max(img, axis=-1)
    minc = np.min(img, axis=-1)
    v = maxc
    delta = maxc - minc

    # Saturation
    s = np.zeros_like(maxc)
    nonzero = maxc > 0
    s[nonzero] = delta[nonzero] / maxc[nonzero]

    # Hue
    h = np.zeros_like(maxc)
    mask = delta > 0

    # 避免除以 0
    r_is_max = (maxc == r) & mask
    g_is_max = (maxc == g) & mask
    b_is_max = (maxc == b) & mask

    h[r_is_max] = ((g[r_is_max] - b[r_is_max]) / delta[r_is_max]) % 6.0
    h[g_is_max] = ((b[g_is_max] - r[g_is_max]) / delta[g_is_max]) + 2.0
    h[b_is_max] = ((r[b_is_max] - g[b_is_max]) / delta[b_is_max]) + 4.0

    h = (h / 6.0) % 1.0  # 正規化到 [0,1)

    hsv = np.stack([h, s, v], axis=-1)
    return hsv

# ---------- HSV -> RGB（純 NumPy） ----------
def hsv2rgb(image):
    '''
    Convert the hsv image into rgb color space and return it.
    image -- HSV image in [0,1], shape (H, W, 3).
    Returns uint8 RGB in [0,255], shape (H, W, 3).
    '''
    hsv = image.astype(np.float64)
    h, s, v = hsv[..., 0], hsv[..., 1], hsv[..., 2]

    h6 = (h * 6.0) % 6.0
    c = v * s
    x = c * (1 - np.abs((h6 % 2) - 1))
    m = v - c

    # 各區段的 RGB'（未加 m 前）
    zeros = np.zeros_like(h)
    rgbp = np.zeros(hsv.shape, dtype=np.float64)

    conds = [
        (0 <= h6) & (h6 < 1),
        (1 <= h6) & (h6 < 2),
        (2 <= h6) & (h6 < 3),
        (3 <= h6) & (h6 < 4),
        (4 <= h6) & (h6 < 5),
        (5 <= h6) & (h6 < 6),
    ]
    choices = [
        np.stack([c, x, zeros], axis=-1),
        np.stack([x, c, zeros], axis=-1),
        np.stack([zeros, c, x], axis=-1),
        np.stack([zeros, x, c], axis=-1),
        np.stack([x, zeros, c], axis=-1),
        np.stack([c, zeros, x], axis=-1),
    ]

    # 根據條件填值
    for cond, choice in zip(conds, choices):
        rgbp[cond] = choice[cond]

    rgb = (rgbp + m[..., None])  # 加回 m，範圍仍在 [0,1]
    rgb = np.clip(np.round(rgb * 255.0), 0, 255).astype(np.uint8)
    return rgb

# ---------- 畫 RGB 直方圖（可沿用你 Task 3 的版本） ----------
from skimage import exposure
def plot_color_histogram(image):
    plt.figure(figsize=(8,5))
    for i, color_name in enumerate(('red','green','blue')):
        hist, bins = exposure.histogram(image[..., i])
        plt.plot(bins, hist, label="RGB"[i], color=color_name)
    plt.title('Color Histogram')
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# ---------- 1) 轉 HSV 並簡單可視化（顯示三個通道） ----------
image_hsv = rgb2hsv(image)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
titles = ['Hue [0..1)', 'Saturation [0..1]', 'Value [0..1]']
for i in range(3):
    axes[i].imshow(image_hsv[..., i], cmap='gray')
    axes[i].set_title(titles[i])
    axes[i].axis('off')
plt.suptitle('HSV Channels')
plt.tight_layout()
plt.show()

# ---------- 2) 在 HSV 中旋轉色相（Hue） ----------
# 例：旋轉 60 度 = 1/6 週期；你也可以改其他角度，例如 120 度=2/6。
hue_shift = 1.0 / 6.0
image_hsv_rot = image_hsv.copy()
image_hsv_rot[..., 0] = (image_hsv_rot[..., 0] + hue_shift) % 1.0

# ---------- 3) 轉回 RGB 並顯示 ----------
image_rot_rgb = hsv2rgb(image_hsv_rot)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image)
axes[0].set_title('Original RGB')
axes[0].axis('off')

axes[1].imshow(image_rot_rgb)
axes[1].set_title(f'Hue-rotated RGB (+{int(hue_shift*360)}°)')
axes[1].axis('off')

plt.tight_layout()
plt.show()

# ---------- 4) 繪製旋轉後 RGB 的色彩直方圖 ----------
plot_color_histogram(image_rot_rgb)
````
